## Extensive Analysis of the Aggregated and cleaned SFU and UBC Datasets

**1. Setup and Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for visualizations
sns.set(style="whitegrid")
```

**2. Load the Cleaned Datasets**

```python
# Load SFU dataset
sfu_df = pd.read_csv("../data/cleaned/SFU_Aggregated_Cleaned.csv")

# Load UBC dataset
ubc_df = pd.read_csv("../data/cleaned/UBC_Aggregated_Cleaned.csv")

# Display the first few rows of each dataset
print("SFU Dataset:")
display(sfu_df.head())

print("UBC Dataset:")
display(ubc_df.head())
```

**3. Exploratory Data Analysis (EDA)**

3.1. Summary Statistics

```python
# Summary statistics for SFU
print("SFU Summary Statistics:")
display(sfu_df.describe())

# Summary statistics for UBC
print("UBC Summary Statistics:")
display(ubc_df.describe())
```

3.2. Participation Trends

```python
# SFU Participation Trends
sfu_participation = sfu_df[["Signed up", "Applied", "Accepted", "Approved", "Realized", "Finished"]].sum()
sfu_participation.plot(kind="bar", title="SFU Participation Trends (2013-2024)")
plt.ylabel("Number of Students")
plt.show()

# UBC Participation Trends
ubc_participation = ubc_df[["Signed up", "Applied", "Accepted", "Approved", "Realized", "Finished"]].sum()
ubc_participation.plot(kind="bar", title="UBC Participation Trends (2013-2024)")
plt.ylabel("Number of Students")
plt.show()
```

3.3. Conversion Rates

```python
# SFU Conversion Rates
sfu_conversion = (
    sfu_df[["Applied", "Accepted", "Approved", "Realized", "Finished"]].sum() 
    / sfu_df["Signed up"].sum()
) * 100
sfu_conversion.plot(kind="bar", title="SFU Conversion Rates (%)")
plt.ylabel("Conversion Rate (%)")
plt.show()

# UBC Conversion Rates
ubc_conversion = (
    ubc_df[["Applied", "Accepted", "Approved", "Realized", "Finished"]].sum() 
    / ubc_df["Signed up"].sum()
) * 100
ubc_conversion.plot(kind="bar", title="UBC Conversion Rates (%)")
plt.ylabel("Conversion Rate (%)")
plt.show()
```

**4. Analyze Trends**

4.1. Higher GPA + Career Motivation == More Funding

```python
# SFU: Funding vs GPA and Motivation
sfu_funding = sfu_df.groupby(["Motivation", "Funding"])["GPA"].mean().unstack()
sfu_funding.plot(kind="bar", title="SFU: Average GPA by Motivation and Funding")
plt.ylabel("Average GPA")
plt.show()

# UBC: Funding vs GPA
ubc_funding = ubc_df.groupby("Funding")["GPA"].mean()
ubc_funding.plot(kind="bar", title="UBC: Average GPA by Funding")
plt.ylabel("Average GPA")
plt.show()
```

4.2. Higher English Proficiency -> Longer Duration of Exchange

```python
# SFU: English Proficiency vs Length of Exchange
sns.scatterplot(data=sfu_df, x="English Proficiency", y="Length of Exchange", hue="Motivation")
plt.title("SFU: English Proficiency vs Length of Exchange")
plt.show()
```

4.3. Career-Motivated Fields

```python
# SFU: Motivation by Background
sfu_motivation = sfu_df.groupby("Backgrounds")["Motivation"].value_counts().unstack()
sfu_motivation.plot(kind="bar", stacked=True, title="SFU: Motivation by Background")
plt.ylabel("Number of Students")
plt.show()

# UBC: Motivation by Background (if available)
if "Motivation" in ubc_df.columns:
    ubc_motivation = ubc_df.groupby("Backgrounds")["Motivation"].value_counts().unstack()
    ubc_motivation.plot(kind="bar", stacked=True, title="UBC: Motivation by Background")
    plt.ylabel("Number of Students")
    plt.show()
```

4.4. Assign Higher Previous Experience to People with Higher GPA

```python
# SFU: GPA vs Prior International Experience
sns.boxplot(data=sfu_df, x="Prior International Experience", y="GPA")
plt.title("SFU: GPA vs Prior International Experience")
plt.show()
```

4.5. High GPA and Career Motivation -> Co-op Before Exchange

```python
# SFU: Co-op Before Exchange vs GPA and Motivation
sfu_coop = sfu_df.groupby(["Motivation", "Co-op Before Exchange"])["GPA"].mean().unstack()
sfu_coop.plot(kind="bar", title="SFU: Average GPA by Motivation and Co-op Before Exchange")
plt.ylabel("Average GPA")
plt.show()
```

**5. Additional Insights**

5.1. Most Popular Backgrounds

```python
# SFU: Top 10 Backgrounds by Participation
sfu_top_backgrounds = sfu_df.groupby("Backgrounds")["Signed up"].sum().nlargest(10)
sfu_top_backgrounds.plot(kind="bar", title="SFU: Top 10 Backgrounds by Participation")
plt.ylabel("Number of Students")
plt.show()

# UBC: Top 10 Backgrounds by Participation
ubc_top_backgrounds = ubc_df.groupby("Backgrounds")["Signed up"].sum().nlargest(10)
ubc_top_backgrounds.plot(kind="bar", title="UBC: Top 10 Backgrounds by Participation")
plt.ylabel("Number of Students")
plt.show()
```

5.2. Funding Distribution

```python
# SFU: Funding Distribution
sfu_funding_dist = sfu_df["Funding"].value_counts()
sfu_funding_dist.plot(kind="pie", autopct="%1.1f%%", title="SFU: Funding Distribution")
plt.show()

# UBC: Funding Distribution
ubc_funding_dist = ubc_df["Funding"].value_counts()
ubc_funding_dist.plot(kind="pie", autopct="%1.1f%%", title="UBC: Funding Distribution")
plt.show()
```

5.3. Campus Distribution

```python
# SFU: Campus Distribution
sfu_campus_dist = sfu_df["SFU Campus"].value_counts()
sfu_campus_dist.plot(kind="pie", autopct="%1.1f%%", title="SFU: Campus Distribution")
plt.show()
```

5.4. Length of Exchange by Motivation

```python
# SFU: Length of Exchange by Motivation
sfu_length_motivation = sfu_df.groupby("Motivation")["Length of Exchange"].mean()
sfu_length_motivation.plot(kind="bar", title="SFU: Average Length of Exchange by Motivation")
plt.ylabel("Average Length (Months)")
plt.show()
```

5.5 Gender Distribution

```python
# SFU: Gender Distribution
sfu_gender_dist = sfu_df["Gender"].value_counts()
sfu_gender_dist.plot(kind="pie", autopct="%1.1f%%", title="SFU: Gender Distribution")
plt.show()

# UBC: Gender Distribution
ubc_gender_dist = ubc_df["Gender"].value_counts()
ubc_gender_dist.plot(kind="pie", autopct="%1.1f%%", title="UBC: Gender Distribution")
plt.show()
```

5.6 Funding and GPA Trends

```python
# SFU: GPA vs Funding
sns.boxplot(data=sfu_df, x="Funding", y="GPA")
plt.title("SFU: GPA Distribution by Funding")
plt.show()

# UBC: GPA vs Funding
sns.boxplot(data=ubc_df, x="Funding", y="GPA")
plt.title("UBC: GPA Distribution by Funding")
plt.show()
```

5.7 Co-op and Exchange Participation

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load SFU dataset
sfu_df = pd.read_csv("../data/cleaned/SFU_Aggregated_Cleaned.csv")

# Calculate acceptance rate for SFU dataset
sfu_df["Acceptance Rate"] = (sfu_df["Accepted"] / sfu_df["Applied"]) * 100

# Calculate the overall average acceptance rate
average_acceptance_rate = sfu_df["Acceptance Rate"].mean()

# Prepare data for the pie chart
data = [average_acceptance_rate, 100 - average_acceptance_rate]
labels = ["Average Acceptance Rate", "Remaining"]

# Plot pie chart for the average acceptance rate
plt.figure(figsize=(8, 8))
plt.pie(
    data, 
    labels=labels, 
    autopct='%1.1f%%', 
    startangle=140, 
)
plt.title("Average Acceptance Rate")
plt.show()
```

```python
# SFU: Co-op Before Exchange vs Participation
sfu_coop_participation = sfu_df.groupby("Co-op Before Exchange")["Signed up"].sum()
sfu_coop_participation.plot(kind="bar", title="SFU: Co-op Before Exchange vs Participation")
plt.ylabel("Number of Students")
plt.show()
```
